Extracting the grid from SciPy's KDTree implmentation
-----------------------------------------------------

SciPy has two KDTree implementations.  One in C and another in raw Python.  Alogrithmically, they are very similar.  In this notebook, we extract the grid from the pure-python version.

```python
%matplotlib inline

import numpy as np
from scipy import spatial
import matplotlib.pyplot as plt
```

First generate some data

```python
N = 2000                                                                        
M = 100                                                                         
r = np.random.normal(0.5, 0.03, N)                                              
t = 2 * np.pi * np.random.rand(N)                                               
points = np.empty((N + M, 2))                                                   
points[:N:, 0] = (r * np.cos(t)).clip(-1, 1)                                    
points[:N, 1] = (r * np.sin(t)).clip(-1, 1)                                     
points[N:, :] = np.random.rand(M, 2)                                            
                                                                                
plt.plot(points[:, 0], points[:, 1], 'r.')
plt.show()
```

```python
def kd_blocks(tree, mins, maxes):
    """Iterate over a KDTree and return the blocks for the leaf nodes
    
    Args:
        tree: tree attribute of a scipy.spatial.KDTree
        mins: minimum values for each dimension
        maxes: maximum values for each dimension
    Returns:
        generator producing:
        mins, maxes, indexes for each leaf node in the KDTree
    """"
    # Iterate over the children nodes with smaller values
    if hasattr(tree, 'less'):                                                   
        # update maxes to reflect the current split
        _maxes = np.array(maxes)                                                
        _maxes[tree.split_dim] = tree.split                                     
        for box in kd_blocks(tree.less, mins, _maxes):                          
            yield box                                                           
    # Iterate over the children nodes with large values
    if hasattr(tree, 'greater'):                                                
        # update maxes to reflect the current split
        _mins = np.array(mins)                                                  
        _mins[tree.split_dim] = tree.split                                      
        for box in kd_blocks(tree.greater, _mins, maxes):                       
            yield box                                                           
                                                                                
    # If we're a leaf node, return data
    if hasattr(tree, 'idx'):                                                    
        minx, miny = mins                                                       
        maxx, maxy = maxes                                                      
        plt.plot([minx, maxx, maxx, minx, minx],                                
                          
        yield mins, maxes, tree.idx     
```

```python
# Make a kd tree with 20 points per leaf
kd = spatial.KDTree(points, leafsize=20)                                        
plt.plot(points[:, 0], points[:, 1], 'r.')

leaf_size = None
for mins, maxes, idx in kd_blocks(kd.tree, kd.mins, kd.maxes):
    minx, miny = mins
    maxx, maxy = maxes
    plt.plot([minx, maxx, maxx, minx, minx],
             [miny, miny, maxy, maxy, miny], 'k')
    leaf_size = max(leaf_size, len(idx))
print leaf_size
```

```python
# Make a kd tree with 50 points per leaf
kd = spatial.KDTree(points, leafsize=50)                                        
plt.plot(points[:, 0], points[:, 1], 'r.')
                                                                                
leaf_size = None
for mins, maxes, idx in kd_blocks(kd.tree, kd.mins, kd.maxes):
    minx, miny = mins
    maxx, maxy = maxes
    plt.plot([minx, maxx, maxx, minx, minx],
             [miny, miny, maxy, maxy, miny], 'k')
    leaf_size = max(leaf_size, len(idx))
print leaf_size
```
